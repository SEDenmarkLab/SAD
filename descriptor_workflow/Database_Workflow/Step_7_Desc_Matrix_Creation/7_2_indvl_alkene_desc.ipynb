```python
import molli as ml
from pprint import pprint
import pandas as pd
import numpy as np

def isolate_all_indvl(full_df: pd.DataFrame, vol_type: str):
    alk_types = ['Mono', 'Gem', 'Cis', 'Gem', 'Trans', 'TriQ2', 'TriQ3', 'TriQ4', 'Tetra']
    for alk in alk_types:
        alk_df = full_df.loc[full_df['Alkene Type'] == alk]
        # print(f'\nFor {alk} this is the dataframe:\n{alk_df}')
        match alk:
            case 'Mono':
                '''
                Q1 | H
                ----------
                H  | H
                '''
                alk_df = alk_df.filter(regex=r'^(?!Q[2,3,4])')
            case 'Gem':
                '''
                Q1 | H
                ----------
                Q4  | H
                '''
                alk_df = alk_df.filter(regex=r'^(?!Q[2,3])')

            case 'Cis':
                '''
                Q1 | Q2
                ----------
                H  | H
                '''
                alk_df = alk_df.filter(regex=r'^(?!Q[3,4])')

            case 'Trans':
                '''
                Q1 | H
                ----------
                H  | Q3
                '''
                alk_df = alk_df.filter(regex=r'^(?!Q[2,4])')

            case 'TriQ2':
                '''
                Q1 | H
                ----------
                Q4 | Q3
                '''
                alk_df = alk_df.filter(regex=r'^(?!Q[2])')

            case 'TriQ3':
                '''
                Q1 | Q2
                ----------
                Q4  | H
                '''
                alk_df = alk_df.filter(regex=r'^(?!Q[3])')

            case 'TriQ4':
                '''
                Q1 | Q2
                ----------
                H  | Q3
                '''
                alk_df = alk_df.filter(regex=r'^(?!Q[4])')

            case 'Tetra':
                '''
                Q1 | Q2
                ----------
                Q4  | Q3
                '''
                pass

            case _:
                raise ValueError(f"Unexpected Value Used: {alk}")

        alk_df.to_csv(f"7_2_{alk}_{vol_type}_Desc.csv")
```

```python
max_vol_df = pd.read_csv("7_1_Full_Alkene_Desc_MaxVol.csv",index_col=0)
isolate_all_indvl(max_vol_df, vol_type='MaxVol')
```

```python
bfs_vol_df = pd.read_csv("7_1_Full_Alkene_Desc_3BFSVol.csv",index_col=0)
isolate_all_indvl(bfs_vol_df, vol_type='3BFSVol')
```
