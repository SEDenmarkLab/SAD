## Current Facial Selectivity Assignments Based on Matched Structures

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def vis_indvl_face(df: pd.DataFrame, alk_type):
    filter_df = df[df['Alkene Type']==alk_type]
    filter_df = format_face_df(filter_df)

    sns.countplot(
        data=filter_df,
        x="Catalyst Designation",
        hue='Addition Face',
        hue_order=['Top', 'Bottom'],
    )

    plt.xlabel('Catalyst ID', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title(f'{alk_type} (n={filter_df.shape[0]}) AD Results', fontsize=16)

    plt.savefig(f'./face_vis/Visualize_{alk_type}.png')
    plt.show()
    plt.close()

def format_face_df(df:pd.DataFrame):

    cat_id_order = {"Alpha":0, "Beta":1}
    facial_addition_order = {'Top':0, 'Bottom':1}

    sort_order = np.argsort(np.select(
        [df['Catalyst Designation'] == 'Alpha'],
        [df['Addition Face'].map(facial_addition_order)],
        default=df['Catalyst Designation'].map(cat_id_order)
    ))

    return df.iloc[sort_order]

def vis_all(df:pd.DataFrame, vol_type:str, to_include:dict):

    num_plots = len(to_include)
    fig, axs = plt.subplots(1, num_plots, figsize=(16,6))

    for alk_type,num in to_include.items():
        filter_df = df[df['Alkene Type']==alk_type]
        filter_df = format_face_df(filter_df)

        ax = axs[num] if num_plots > 1 else axs

        sns.countplot(
            data=filter_df,
            x="Catalyst Designation",
            hue='Addition Face',
            hue_order=['Top', 'Bottom'],
            ax=ax
        )

        ax.set_xlabel('AD Mixture', fontsize=18)
        ax.set_ylabel('Frequency', fontsize=18)
        ax.set_ylim(bottom=0, top=150)
        ax.set_title(f'{alk_type} (n={filter_df.shape[0]}) AD Results', fontsize=18)
        ax.set_xticklabels([r'$\alpha$', r'$\beta$'])
        ax.tick_params(axis='x', labelsize=18)
        ax.tick_params(axis='y', labelsize=16)
        ax.legend(fontsize=16)
        if num != 0:
            ax.get_legend().remove()
        
    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
bfsvol_face_df = pd.read_csv("6_2_Diol_Selectivity_3BFSVol_Desc.csv")

bfsvol_small_df = bfsvol_face_df[['Product ID', 'Catalyst Designation', 'Alkene Type', 'Addition Face']]

custom_order1 = {
    'Mono':0,
    'Gem':1,
    'Cis':2,
    'Trans':3,
}
custom_order2 = {
    'TriQ2':0,
    'TriQ3':1,
    'TriQ4':2,
    'Tetra':3
}

print("BFS")
vis_all(bfsvol_small_df, '3BFS Volume', custom_order1)
vis_all(bfsvol_small_df, '3BFS Volume', custom_order2)
```

```python
maxvol_face_df = pd.read_csv("6_2_Diol_Selectivity_MaxVol_Desc.csv")
maxvol_small_df = maxvol_face_df[['Product ID', 'Catalyst Designation', 'Alkene Type', 'Addition Face']]

#Isolates the indexes  of the maxvol dataframe to be equal to that of the BFSVol index for a one to one comparison
maxvol_small_df = maxvol_small_df.loc[maxvol_face_df['Product ID'].isin(bfsvol_small_df["Product ID"])]
print("MaxVol")
vis_all(maxvol_small_df, 'Max Volume', custom_order1)
vis_all(maxvol_small_df, 'Max Volume', custom_order2)
```
